```python
session1_label = [1,2,3,0,2,0,0,1,0,1,2,1,1,1,2,3,2,2,3,3,0,3,0,3]
session2_label = [2,1,3,0,0,2,0,2,3,3,2,3,2,0,1,1,2,1,0,3,0,1,3,1]
session3_label = [1,2,2,1,3,3,3,1,1,2,1,0,2,3,3,0,2,3,0,0,2,0,1,0]

emotions_dict = {0: "NEUTRAL", 1:"SAD", 2:"FEAR", 3:"HAPPY" }
trial_emotion_1 = [emotions_dict[x] for x in session1_label]
trial_emotion_2 = [emotions_dict[x] for x in session2_label]
trial_emotion_3 = [emotions_dict[x] for x in session3_label]

n_receptors = 62
```

```python
import scipy.io

def read_mat(_path, _name, _labels, _output_path):

    # Load .mat file
    mat_file = scipy.io.loadmat(_path + "/" + _name)

    variable_names = [var for var in mat_file if not var.startswith('__')] 

    for i, name in enumerate(variable_names):

        trial_signals = []
        
        for j in range(n_receptors):
            trial_signals.append(mat_file[name][j])

        trial_signals = np.array(trial_signals).T

        df = pd.DataFrame(trial_signals, columns=channels["channels"])

        #Stores the data in a parquet file
        new_file_name = f"{_name.rstrip('.mat')}_{i}_signals.parquet"
        df.to_parquet(f"{_output_path}/{_labels[i]}/{new_file_name}", compression='gzip')
```

```python
def read_folder_data(_path, _output_path, _labels):
    #Gets the file names in the folder and selects just those with .mat extension
    file_names = os.listdir(_path)
    mat_names = tuple(filter(lambda s: ".mat" in s, file_names))
    for name in mat_names:
        read_mat(_path, name, _labels, _output_path)
```

```python
read_folder_data("SEED_IV/1", "SEED_IV_OUTPUT/1", trial_emotion_1)
read_folder_data("SEED_IV/2", "SEED_IV_OUTPUT/2", trial_emotion_2)
read_folder_data("SEED_IV/3", "SEED_IV_OUTPUT/3", trial_emotion_3)
```
